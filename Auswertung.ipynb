```python
#HIDDEN
from ipywidgets import *
from IPython.display import display, clear_output, Image
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Semikolon durch Komma ersetzen
def replace_semi(datei):
    f = open(datei, "r")
    content = f.read()
    f.close()
    content = content.replace(";", ",")
    f = open(datei, "w")
    f.write(content)
    f.close()
    return
```

```python
# Ergebnisabbildung
def print_result(fläche, volumen): #hier noch den rest ergänzen
    # Hier den Abstand der Profile eintragen
    weglänge = [5,20] 
    
    plt.figure(figsize=(15,8))
    plt.plot(weglänge,fläche,'-', markersize=10)
    plt.plot(weglänge,volumen,'.', markersize=10)
    plt.grid()
    plt.xlabel('Weglänge [m]');
    plt.ylabel('Abweichung');
    plt.title('Abweichung von Strecke/Fläche/Volumen in abhängigekit zur zurückgelegten Strecke');
    plt.legend(['Fläche','Volumen'])
    #plt.legend(['Strecke', 'Fläche','Volumen'])
    plt.show()
    #Hier die Einheiten beahcten!!!
```

```python
# Differenz berechnen
def delta(soll,ist):
    delta = []
    for i in range(len(soll)):
        delta.append(ist[i]-soll[i]) #wenn ist > soll, ist das delta positiv
    #print('Delta =', delta)
    return (delta)
```

```python
# Variable aus File entnehmen
def get_variable(data, variable1, variable2):
    var = []
    for index, row in data.iterrows():
        if row['Element'] == variable1:
            var.append(row[variable2])
    #print('Variable =', var)
    return var
```

```python
# Matrix aus den Punkten erstellen für Streckenberechnung

#profile = Anzahl der Profile
#point = Anzahl Punkte je Profil
#vari = Anzahl Variablen, hier x,y,z
def get_points(data):
    profile, point, vari = (2, 4, 3)
    arr = [[[0 for i in range(vari)] for j in range(point)] for d in range(profile)]
    
    #Liste mit den x,y,z Werten aller Punkte
    ar = []
    for index, row in data.iterrows():
            if row['Element'] == 'Punkt1' or row['Element'] == 'Punkt2' or row['Element'] == 'Punkt3' or row['Element'] == 'Punkt4':
                ar.append((row['X'], row['Y'], row['Z']))

    #Matrix belegen            
    for i in range(profile):
        for j in range(point):
            for k in range(vari):
                arr[i][j][k] = ar[j + (i*4)][k]

    #print(ar,'\n')
    #print(arr)
    #print(arr[0][1][1]) #Profil 1 Punkt 2 Variable y
    return
```

```python
#das passiert, wenn man auf Button klickt
def start(x):
    clear_output()
    display(referenz_file, test_file, button_import)
    
    replace_semi(referenz_file.value)
    replace_semi(test_file.value)
    
    referenz_data = pd.read_csv(referenz_file.value)
    test_data = pd.read_csv(test_file.value)
    print('Referenzdaten:')
    display(referenz_data)
    print('Testdaten:')
    display(test_data)
        #print('Referenzdaten: \n\n',referenz_data,'\n\nTestdaten: \n\n',test_data)
    
    #Punkte zu Strecken
    points_soll = get_points(referenz_data)
    points_ist = get_points(test_data)
    
    #Fläche
    fläche_soll = get_variable(referenz_data, 'Fläche', 'Fläche')
    fläche_ist = get_variable(test_data, 'Fläche', 'Fläche')
    fläche_delta = delta(fläche_soll,fläche_ist)
    
    #Volumen
    volumen_soll = get_variable(referenz_data, 'Volumen', 'Volumen')
    volumen_ist = get_variable(test_data, 'Volumen', 'Volumen')
    volumen_delta = delta(volumen_soll,volumen_ist)
    
    print()
    print_result(fläche_delta, volumen_delta)
```
