```python
#HIDDEN
from ipywidgets import *
from IPython.display import display, clear_output, Image
import pandas as pd
```

```python
referenz_file = widgets.Text(description = 'Referenz: ', layout = Layout(height='auto', width='auto'))
test_file = widgets.Text(description = 'Test: ', layout = Layout(height='auto', width='auto'))
button_import = widgets.Button(description = 'Importieren', button_style = '', layout=Layout(height='auto', width='100%'))
display(referenz_file, test_file, button_import)

def start(x):
    global referenz_data
    referenz_data = pd.read_csv(referenz_file.value)
    global test_data
    test_data = pd.read_csv(test_file.value)
    print('Referenzdaten: \n\n',referenz_data.head(),'\n\nTestdaten: \n\n',test_data.head())

button_import.on_click(start)
```
